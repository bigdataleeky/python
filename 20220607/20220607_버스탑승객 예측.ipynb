```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.displot(traindata,x = '18~20_ride',bins=50)
plt.axis([0,50,0,400000])
plt.title('퇴근 시간 승차인원 히스토그램')
plt.show()
```

```python
traindata.groupby('18~20_ride').size().head()
```

```python
traindata['date'] =  pd.to_datetime(traindata['date'])
```

```python
# 날자를 문자형 요일로 추출해서 저장
traindata['weekday']  = traindata['date'].dt.strftime('%a')
```

```python
import numpy as np
np.unique( traindata['weekday'])
```

```python
def changeWeek(x):
    if x == 'Sun':
        return '일'
    elif x == 'Mon':
        return '월'
    elif x == 'Tue':
        return '화'
    elif x == 'Wed':
        return '수'
    elif x == 'Thu':
        return '목'
    elif x == 'Fri':
        return '금'
    elif x == 'Sat':
        return '토'    

traindata['weekday_kor'] = traindata['weekday'].apply(changeWeek)
traindata.tail()
```

```python
#요일별 퇴근시간 평균 탑승객 수
sns.barplot(data=traindata,x='weekday_kor', y='18~20_ride')
plt.title("요일에따른 퇴근시간 평균 탑승객 수")
plt.show()
```

```python
# 버스종류별 퇴근시간 평균 탑승객 수
sns.barplot(data = traindata, x='in_out',y='18~20_ride')
plt.title("버스 종류에 따른 평균 퇴근시간대 탑승객 수")
plt.show()
```

```python
#일별 출퇴근 시간의 총 승차 인원
print(traindata.columns)
traindata['8~10_ride'] =  traindata['8~9_ride'] + traindata['9~10_ride']
traindata.head()
```

```python
traindata.groupby('date')[['18~20_ride', '8~10_ride' ] ].sum()
```

```python
traindata.groupby('date')[['18~20_ride', '8~10_ride' ] ].sum().reset_index()
```

```python
# 일별 출퇴근 시간 탑승객 수 시각화
day_ride_820 =  traindata.groupby('date')[['18~20_ride', '8~10_ride' ] ].sum().reset_index()
plt.plot('date','18~20_ride', data = day_ride_820, label="퇴근시간 탑승객수" )
plt.plot('date','8~10_ride',data = day_ride_820, label="출근시간 탑승객수")
plt.gcf().autofmt_xdate() # x축의 라벨이 서로 겹치지 않도록 설정
plt.legend(loc = 0) # 그래프 상에서 최적의 위치에 범례표시
plt.title('일별 출퇴근 시간 탑승객 수')
plt.show()
```
