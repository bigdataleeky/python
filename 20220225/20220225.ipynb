```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(
    {'A':[14.00,90.20,90.95,96.27,91.21],
     'B':[103.02,107.26,110.35,114.23,114.68], 
     'C':['big','small','big','small','small']})

df
```

```python
# 평균, 최대 최소 값을 이용해서 scaling
df["A"] - df["A"].mean()
```

```python
df["A"].max() - df["A"].min()
```

```python
# (x - min(x) ) / ( max(x) - min(x) )
(df["A"] - df["A"].mean()) / (df["A"].max() - df["A"].min())
```

```python
def scale1(x):
    return (x - x.mean()) / (x.max() - x.min())    
```

```python
scale1(df["A"])
```

```python
scale1(df["B"])
```

```python
# 범주형은 one hot encoding
pd.get_dummies(df['C'])
```

```python
df["A"] = scale1(df["A"])
df["B"] = scale1(df["B"])
df
```

```python
temp = pd.get_dummies(df['C'])
df= pd.concat([df,temp],axis=1)
df = df.drop(["C"], axis=1)
df
```

```python
df = pd.DataFrame(
    {'A':[14.00,90.20,90.95,96.27,91.21],
     'B':[103.02,107.26,110.35,114.23,114.68], 
     'C':['big','small','big','small','small']})

df
```

```python
#z-score  
# (x - x.mean() ) / (x.std())
def scale2(x):
    return (x - x.mean() ) / (x.std())
```

```python
df["A"] = scale2(df["A"])
df["B"] = scale2(df["B"])
df
```

```python
temp = pd.get_dummies(df["C"])
df = df.iloc[:,:-1]
df = pd.concat([df,temp],axis=1)
df
```
