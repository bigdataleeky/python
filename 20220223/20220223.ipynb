```python
x_1 = range(100)
y_1 = [np.cos(value) for value in x_1]
y_2 = [np.sin(value) for value in x_1]
plt.plot(x_1,y_1)
plt.plot(x_1,y_2)
plt.show()
```

```python
#figure  axes
fig, ax = plt.subplots()
x_1 = range(100)
y_1 = [np.cos(value) for value in x_1]
ax.plot(x_1,y_1)
ax.set(title='cos graph',xlabel='x', ylabel='y')
plt.show()
```

```python
fig = plt.figure() # figure 반환
fig.set_size_inches(10,10)  # figure 크기 지정
ax_1 = fig.add_subplot(1,2,1) # 첫번째 그래프 생성
ax_2 = fig.add_subplot(1,2,2) # 두번째 그래프 생성
ax_1.plot(x_1,y_1, c='b')
ax_2.plot(x_1,y_2, c='g')
plt.show()
```

```python
x = np.linspace(-1,1,100)
y_1 = np.sin(x)
y_2 = np.cos(x)
y_3 = np.tan(x)
y_4 = np.exp(x)
fig, ax =  plt.subplots(2,2)
ax[0,0].plot(x,y_1)
ax[0,1].plot(x,y_2)
ax[1,0].plot(x,y_3)
ax[1,1].plot(x,y_4)
plt.show()
```

```python
x_1 = range(100)
y_1 = range(100)

x_2= range(100)
y_2 = [value+100 for value in x_2]

plt.plot(x_1,y_1,c='r', linestyle = 'dashed')
plt.plot(x_2,y_2,c='b', linestyle = 'dotted')
plt.title("Tow lines")
plt.show()
```

```python
# 범례  legend
# shadow 매개변수로 그림자 효과
# loc 매개변수로 위치
    # center upper right... 총 11
    # best 최적의 위치에 자동으로 설정
```

```python
x_1 = range(100)
y_1 = range(100)

x_2= range(100)
y_2 = [value+100 for value in x_2]

plt.plot(x_1,y_1,c='r', linestyle = 'dashed',label='line_1')
plt.plot(x_2,y_2,c='b', linestyle = 'dotted',label='line_2')
plt.legend(
    shadow=True,
    fancybox = False,
    loc = 'best'
)
plt.title("Tow lines")
plt.xlabel('x_line')
plt.ylabel('y_line')
```
