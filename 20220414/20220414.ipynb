```python
# 퍼셉트론 학습
# 테이터와 함께 데이터를 인식하는 퍼셉트론(가중치)이 주어짐
# OR 데이터는 2차원 특징 벡터--> 우리가 수동으로 가중치 계산이 가능
# skearn 필기 숫자  64차원 특징 벡터,,  셈플이 1797개.. 수동계산이 불가능
# --- > 알고리즘이 필요
```

```python
'''
쉐도우코드
수영학습
1.적절한 동작을 취한다.
while(true)
    동작에 따라 수영을 하고 평가를 한다
    if(만족) break
    더 나은 방향으로 수영을 하도록 수정한다.
동작을 기억한다    
'''
```

```python
'''
초기 벡터 w = (팔 돌기는 속도, 각도, 몸에서의 거리..) 초기화
while(true)
    w에 따라 수영을 하고 동작을 w의 점수 j(w)를 계산 -- 손실함수(보통 오류의 양)
    if(j(w) 만족스러움)break
    더 나은 방향으로 w+ 를 계산(미분을 사용)
    w = w +  w+
w를 기억    
'''
```

```python
# 손실함수가 갖춰어야할 조건
# w가 훈련 집합에 있는 샘플을 모두 맞추면 , 정확률이 100%이면 J(w)는 0이다
# w가 틀리는 샘플이 많을수록 J(w)값은 커진다
```
