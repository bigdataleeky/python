```python
#fnWeight = lambda x: int(re.findall('\d+', x.split('/')[1])[0])
# regular_df['height/weight'].apply(fnWeight)
```

```python
regular_df['weight'] =  [int(re.findall('\d+', i.split('/')[1])[0]) if pd.notna(i) else i  for i in  regular_df['height/weight']]
regular_df.head()
  
```

```python
regular_df['height'] =  [int(re.findall('\d+', i.split('/')[0])[0]) if pd.notna(i) else i  for i in  regular_df['height/weight']]
regular_df.head()
```

```python
regular_df['wegith_per_height']  = regular_df['weight'] / regular_df['height']
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
# 정규시즌과 프리시즌의 상관관계
correlation =  regular_df['wegith_per_height'].corr(regular_df['OBP'])  # OBP 출류율
sns.scatterplot(x = regular_df['wegith_per_height'], y = regular_df['OBP'])
plt.title("몸무게와 키 그리고 OBP의 상관관계 " + str(correlation))
plt.xlabel("몸무게/키")
plt.ylabel("OBP")

plt.subplot(1,2,2)
correlation =  regular_df['wegith_per_height'].corr(regular_df['SLG'])  # SLG 장타율
sns.scatterplot(x = regular_df['wegith_per_height'], y = regular_df['SLG'])
plt.title("몸무게와 키 그리고 SLG의 상관관계 " + str(correlation))
plt.xlabel("몸무게/키")
plt.ylabel("SLG")

plt.show()
```
