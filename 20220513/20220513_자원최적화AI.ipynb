```python
# RNN  15    
daily_data = df2[['15']]
daily_data.head()
```

```python
#   15  window feature       
n_times_advance = 1
n_times_windows = 168
```

```python
#   window features    
for k in range(n_times_advance,n_times_advance+n_times_windows):
    daily_data['15-%d'%k] = np.zeros(len(daily_data['15']))
daily_data.head()    
```

```python
#   column  0     
# len(daily_data['15'])
for i in range(n_times_advance+n_times_windows,len(daily_data['15'])):  # 169-6168
    for j in range(n_times_advance,n_times_windows+n_times_advance):
        daily_data['15-%d'%j][i] = daily_data['15'][i-j]
```

```python
daily_data2 =  daily_data.iloc[n_times_advance+n_times_windows:]
daily_data2.head()
```

```python
#    
daily_data2.to_csv('daily_data_seasonality_168.csv',index=True)
```

```python
#    
daily_data3 =  pd.read_csv('daily_data_seasonality_168.csv',index_col=0)
daily_data3.head()
```

```python
daily_data3.index
```

```python
#  /   
#from sklearn.model_selection import train_test_split
#        
# 21 9 1 ~ 9 14 15    
# 24*14 = 336  ( 2  )
# 336       
```

```python
length_train_168 = len(daily_data3)-336
length_train_168
```

```python
#         0      
#  
# Standard Scaler :  Feature  0  1 ...     .
   #   .()
# RobustScaler :   median quartile()  -> 
# MinMax Scaler :  Feature 0 1 
# Normalizer :      ..   row    1

# MinMax     
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler_168 =  MinMaxScaler(feature_range=(0,1))
scaler_168
```

```python
daily_data_sc_168= scaler_168.fit_transform(daily_data3)
daily_data_sc_168
```

```python
data_train_168 =  daily_data_sc_168[:length_train_168]
data_test_168 = daily_data_sc_168[length_train_168:]
len(data_train_168), len(data_test_168)
```

```python
len(data_train_168[0]),len(data_train_168)
```

```python
data_train_Y_168 = data_train_168[:,0]
data_train_X_168 = data_train_168[:,1:]
data_test_Y_168 = data_test_168[:,0]
data_test_X_168 = data_test_168[:,1:]
```

```python
# RNN     3   
data_train_X_168.shape
# window features 168
# (h) 7 
#   24  
#(5663,24,7)
data_train_X_168 =  data_train_X_168.reshape(length_train_168,24,7)
```

```python
#    
import keras
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense,SimpleRNN,Flatten   # SimpleRNN   
```

```python
model_168 = Sequential()
model_168.add(SimpleRNN(50,input_shape=(24,7)))
model_168.add(Flatten())
model_168.add(Dense(1))
opt = tf.keras.optimizers.Adam(learning_rate=0.0001)
model_168.compile(loss="mean_squared_error",optimizer = opt)
model_168.fit(data_train_X_168,data_train_Y_168,epochs=500)
model_168.summary()
```

```python
print(data_test_X_168.shape)
preds_168 = model_168.predict(data_test_X_168.reshape(336,24,7))
```

```python
#       axis = 1    
data_predicted_168 = np.concatenate((preds_168,data_test_X_168),axis = 1)
data_predicted_168
```
