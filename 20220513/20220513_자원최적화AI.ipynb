```python
# 시계열 데이터 학습은 window Feature를 생성
# 시간단위... 1시간단위로 학습, 1일 단위로 학습, 1주일 단위로 학습 결정
# 여기서는 1주일 단위로 24시간으로 구성 7*24 = 168 window를 설정할수 있음
# window Feature를 168로 설정한다.
```

```python
# 데이터 불러오기....
# pd.read_csv('daily_data_seasonality.csv')
```

```python
# RNN을 이용해서 15분 최대전력을 예측하는 것을 보여줌
daily_data = df2[['15분']]
daily_data.head()
```

```python
# 각 시간대의 15분 최대수요전력으로 window feature와 미리 예측하려는 시간 범위를 설정  
n_times_advance = 1
n_times_windows = 168
```

```python
# 학습에 이용할 window features 변수를 생성  
for k in range(n_times_advance,n_times_advance+n_times_windows):
    daily_data['15분-%d'%k] = np.zeros(len(daily_data['15분']))
daily_data.head()    
```

```python
# 새롭게 추가한 column의 값에 0대신 실제 해당하는 최대수요전력의 값을 할당해야함
# len(daily_data['15분'])
for i in range(n_times_advance+n_times_windows,len(daily_data['15분'])):  # 169-6168
    for j in range(n_times_advance,n_times_windows+n_times_advance):
        daily_data['15분-%d'%j][i] = daily_data['15분'][i-j]
```

```python
daily_data2 =  daily_data.iloc[n_times_advance+n_times_windows:]
daily_data2.head()
```

```python
# 생성된 학습용 데이터 저장하기
daily_data2.to_csv('daily_data_seasonality_168.csv',index=True)
```

```python
# 생성된 학습용 데이터 불러오기
daily_data3 =  pd.read_csv('daily_data_seasonality_168.csv',index_col=0)
daily_data3.head()
```

```python
daily_data3.index
```

```python
# 훈련 / 테스트 데이터 분리
#from sklearn.model_selection import train_test_split
# 날자 정보를 이용해서 훈련 및 평가 데이터를 분리한다
# 21년 9월 1일 ~ 9월 14일까지의 15분 최대 수요전력을 예로 든다
# 24*14 = 336개의 최대수요전력을 예측함(마지막 2주 데이터를 정답)
# 336개를 제외한 모든 날짜 데이터가 훈련 데이터로 설정
```

```python
length_train_168 = len(daily_data3)-336
length_train_168
```

```python
# 데이터 값이 너무 크거나 작으면 알고리즘을 이용한 학습에서 0으로 되거나 또는 무한대로 결과나 나올수 있다
# 스케일링 필요
# Standard Scaler : 각 Feature의 평균을 0 분산을 1로 설정... 모든 특성들이 같은 스케일을 가진다.
   # 이상치를 제거해야 한다.(존재한다면)
# RobustScaler : 평균과 분산대신에 median과 quartile(사분위수)을 사용 ->이상치의 영향을안받는다
# MinMax Scaler : 모든 Feature가 0과 1사이에 위치한다
# Normalizer : 위의 방법들은 각 컬럼들의 통계치를 이용하는데.. 이 방식은 row를 기준으로 정규화 유클리드거리가 1이되도록

# MinMax를 사용함    
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler_168 =  MinMaxScaler(feature_range=(0,1))
scaler_168
```

```python
daily_data_sc_168= scaler_168.fit_transform(daily_data3)
daily_data_sc_168
```

```python
data_train_168 =  daily_data_sc_168[:length_train_168]
data_test_168 = daily_data_sc_168[length_train_168:]
len(data_train_168), len(data_test_168)
```

```python
len(data_train_168[0]),len(data_train_168)
```

```python
data_train_Y_168 = data_train_168[:,0]
data_train_X_168 = data_train_168[:,1:]
data_test_Y_168 = data_test_168[:,0]
data_test_X_168 = data_test_168[:,1:]
```

```python
# RNN 의 데이터로 사용하기 위해서는 3차원 텐서가 되어야 함
data_train_X_168.shape
# window features는 168
# 시퀀스(h) 7 
# 각 시퀀스는 24개의 데이터 가짐
#(5663,24,7)
data_train_X_168 =  data_train_X_168.reshape(length_train_168,24,7)
```

```python
# 모델 구축 및 훈련
import keras
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense,SimpleRNN,Flatten   # SimpleRNN 단순 신경망을 사용
```

```python
model_168 = Sequential()
model_168.add(SimpleRNN(50,input_shape=(24,7)))
model_168.add(Flatten())
model_168.add(Dense(1))
opt = tf.keras.optimizers.Adam(learning_rate=0.0001)
model_168.compile(loss="mean_squared_error",optimizer = opt)
model_168.fit(data_train_X_168,data_train_Y_168,epochs=500)
model_168.summary()
```

```python
print(data_test_X_168.shape)
preds_168 = model_168.predict(data_test_X_168.reshape(336,24,7))
```

```python
# 예측한 데이터와 테스트 데이터를 연결  axis = 1을 이용해서 좌에서 우로 연결
data_predicted_168 = np.concatenate((preds_168,data_test_X_168),axis = 1)
data_predicted_168
```
